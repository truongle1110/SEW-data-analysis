```python
from datetime import datetime, timedelta
import numpy as np

def count_weekends(start_date, end_date):
    """
    Count the number of Saturdays and Sundays between two dates.

    Args:
        start_date (str): Start date in the format 'YYYY-MM-DD'.
        end_date (str): End date in the format 'YYYY-MM-DD'.

    Returns:
        int: Number of weekend days between the two dates.
    """
    # Convert string dates to datetime objects
    start = datetime.strptime(start_date, "%Y-%m-%d")
    end = datetime.strptime(end_date, "%Y-%m-%d")

    # Initialize count
    weekend_count = 0

    # Loop through each date in the range
    current_date = start
    while current_date <= end:
        # Check if the current date is a Saturday (5) or Sunday (6)
        if current_date.weekday() in (5, 6):  # 5 = Saturday, 6 = Sunday
            weekend_count += 1
        current_date += timedelta(days=1)

    return weekend_count

# Example usage
dates = ["2023-09-05","2023-09-08","2023-10-09","2023-10-10","2023-10-13","2023-10-16","2023-10-17","2023-10-17","2023-10-18","2023-10-19","2023-10-20","2023-10-26","2023-11-10","2023-11-21","2024-01-09","2024-01-11","2024-01-11","2024-01-15","2024-01-18","2024-01-23","2024-01-26","2024-02-02","2024-02-02","2024-02-03","2024-03-14","2024-03-19","2024-04-15","2024-04-29","2024-04-29","2024-05-15","2024-05-16","2024-05-16","2024-05-17","2024-05-17","2024-05-29","2024-05-30","2024-06-04","2024-06-05","2024-06-17","2024-07-26","2024-08-20"]
#start_date = "2023-10-13"
#end_date = "2023-10-16"

TBOs = []
for index, i in enumerate(dates):
    if index<len(dates)-1:
        start_date = i
        end_date = dates[index+1]
    
        weekends = (count_weekends(start_date, end_date))/2
        print(f"Number of weekends between {start_date} and {end_date}: {weekends}")
        TBOs.append(weekends)
print(np.array(TBOs))
```
